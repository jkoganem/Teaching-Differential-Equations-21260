## Boundary value problems

If a differential equation is coupled with conditions specified at two different points, then we refer to these conditions as **boundary condition**s (as opposed to initial conditions) and the associated problem as a **boundary value problem (BVP)**. Many physical systems are naturally modeled by boundary value problems. 

A prototype would be a 2nd order problem of the form 
$$
\begin{cases}
    y''(x) + p(x)y'(x) + q(x)y(x) = f(x), \; x \in (a,b) \\
    y(a) = 0, \; y(b) = 0,
\end{cases}
$$
where $a < b$ and $p,q: \mathbb{R} \to \mathbb{R}$ are continuous. In contrast with initial value problems, a boundary value problem (BVP) may admit one unique solution, infinitely many solutions, or no solutions even if the coefficients $p, q: \mathbb{R} \to \mathbb{R}$ are continuous functions over the interval $(a,b)$. The most common types of boundary conditions are given in the following table.
$$
\begin{array}{ |c|c|c| }
 \hline
 \text{Type of BC} & \text{Homogeneous BC} & \text{Inhomogeneous BC} \\ \hline
 \text{Dirichlet} & y(a) = 0 & y(a) = \alpha, \alpha \neq 0  \\
 \text{Neumann} & y'(a) = 0 & y'(a) = \alpha, \alpha \neq 0 \\
 \text{Robin} & \beta y(a) + \gamma y'(a) = 0 & \beta y(a) + \gamma y'(a) = \alpha, \alpha \neq 0 \\
 \hline
\end{array}
$$


### Eigenvalue problems
A boundary value problem of the type 
$$
\begin{cases}
    y''(x) + p(x)y'(x) +\lambda q(x)y(x) = 0, \; x \in (a,b) \\
    \alpha y(a) + \beta y'(a) = 0, \; \gamma y(b) + \tau y'(b) = 0,
\end{cases}
$$
where $\lambda \in \mathbb{R}$ is an unspecified parameter is an example of an **eigenvalue problem**. Any $\lambda \in \mathbb{R}$ for which the equation admits a non-zero solution subject to the given boundary conditions is called an **eigenvalue** of the problem, and the associated solutions are called **eigenfunctions** associated to the eigenvalue $\lambda$. 

**Remark:** The abstract way to write this is to introduce the linear operator $L: C^2((a,b) ; \mathbb{R}) \to C^0((a,b) ; \mathbb{R})$ defined via $L(y) = y'' + p y'$. Then the ODE for the eigenvalue problem can be written as
$$
L(y) = - \lambda q y.
$$
From this formulation we can see where the terminology comes from. In linear algebra, one is often interested in finding non-zero vectors $v$ such that $Av = \lambda v$ for some square matrix $A$. This is a generalized version of the eigenvalue problem for matrices applied to the linear operator $L$ acting on functions. 

Many problems in physics can be reduced to the study of eigenvalue problems. For example, the time-independent SchrÃ¶dinger equation in quantum mechanics is an eigenvalue problem. The eigenvalues correspond to the allowed energy levels of a quantum system, and the eigenfunctions correspond to the wave functions of the system at those energy levels. 

Note that since the given problem is a homogeneous problem, in the sense that both the equation and the boundary conditions are of homogeneous type, any scalar multiple of an eigenfunction is also an eigenfunction. Also, under mild conditions on the coefficients $p,q$, one can show that two eigenfunctions associated to two different eigenvalues must be linearly independent. 

> **Example** Let $\mathbb{R} \ni L > 0$. Consider the eigenvalue problem 
> $$
> \begin{cases}
> 	y''(x) + \lambda y(x) = 0, \; x \in (0,L) \\
> 	y(0) = y(L) = 0.
> \end{cases}
> $$
> We consider the following cases.
> 
> 1. If $\lambda = 0$, then
> $$
> y''(x) =0 \implies y'(x) = c_1 \implies y(x) = c_1x + c_2, \; x \in (0,L)
> $$
> for some constants $c_1, c_2$. By using the boundary condition on $y(x)$, we find that 
> $$
> y(0) = c_2 = 0 \\
> y(L) = c_1L = 0 \implies c_1 = 0.
> $$
> Thus the unique solution satisfying the equation and the boundary conditions is the zero solution, so zero is not an eigenvalue of the eigenvalue problem. 
> 
> 2. If $\lambda < 0$, then 
> $$
> y(x) = c_1 e^{ -\sqrt{-\lambda} x } + c_2 e^{\sqrt{-\lambda} x}, \; x \in (0,L)
> $$
> By using the boundary conditions on $y(x)$, we find that
> $$
> y(0) = c_1 + c_2 = 0 \implies c_2 = -c_1 \\
> y(L) = c_1 e^{ -\sqrt{-\lambda} L } + c_2 e^{\sqrt{-\lambda} L} = 0.
> $$
> Thus, we must have 
> $$
> 0 = c_1 e^{ -\sqrt{-\lambda} L } - c_1 e^{\sqrt{-\lambda} L} = c_1  \left( e^{ -\sqrt{-\lambda} L } - e^{\sqrt{-\lambda} L} \right).
> $$
> This implies that $c_1 = c_2 = 0$, so once again, the unique solution satisfying the equation is the zero solution, so the eigenvalue problem does not admit any negative eigenvalues. 
> 
> 3 . If $\lambda > 0$, then 
> $$
> y(x) = c_1 \cos(\sqrt{\lambda} x) + c_2 \sin(\sqrt{\lambda} x), \; x \in (0,L).
> $$
> The boundary conditions require
> $$
> 0 = y(0) = c_1 
> $$
> and 
> $$
> 0 = y(L) = c_2 \sin( \sqrt{\lambda} L ).
> $$ 
> Since we are interested in identifying non-zero solutions to the eigenvalue problem, we may assume that $c_2 \neq 0$. Then we see that the boundary condition is satisfied if 
> $$
>  \sqrt{\lambda } L = n \pi, \; n \in \mathbb{Z} \setminus \{0\}.
> $$
> Thus we can parametrize the eigenvalues with positive integers and 
> $$
>  \lambda_n = \left( \frac{n \pi}{L} \right)^2, \; n \in \mathbb{Z}^+
> $$
> are the eigenvalues of the problem. The corresponding eigenfunctions (up to a multiplicative scalar) of the problem are
> $$
> y_n(x) = \sin \left( \frac{n \pi}{L}  x\right), \; x \in (0,L).
> $$
