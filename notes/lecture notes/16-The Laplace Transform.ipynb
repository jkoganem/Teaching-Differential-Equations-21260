## The Laplace transform

**Definition**
> We define the *Laplace transform* of a function $ f: [0,\infty) \to \mathbb{R} $ to be 
> $$
> \mathcal{L}\{f\} (s) := \int_0^\infty e^{-st} f(t) \; dt
> $$
> for all values $ s \in \mathbb{R} $ for which the improper integral converges.

Before we proceed we make the following remarks.

**Remark:** 

1. Here $ \mathcal{L}\{ f \}(s) $ is a function of $ s $: the Laplace transform takes in a function $ f $ as an input and gives back a function of $ s $, $ F(s) := \mathcal{L}\{f \} (s) $, where $ s $ is real-valued.

2. We often adopt the following abuse of notation: one, we sometimes omit the $ s $ and simply write $ \mathcal{L} \{f\} $ instead. Two, we also sometimes write $ \mathcal{L} \{f(t)\} $ even though it is more appropriate to write $ \mathcal{L} \{f(\cdot)\} $.

3. The integral appearing in the definition of the Laplace transform is an improper integral:
$$
\int_0^\infty e^{-st} f(t) \; dt = \lim\limits_{\gamma \to \infty} \int_0^\gamma e^{-st} f(t) \; dt.
$$
The improper integral may not be well-defined for a particular value of $ s $ and $ f $ (i.e. the integral may not converge). In general, the Laplace transform is only well-defined for certain values of $ s $; in the section to follow we give sufficient conditions to guarantee the existence of the Laplace transform.

4. The main utility of the Laplace transform for us is that it allows us to solve inhomogeneous linear equations where the forcing function on the right-hand side is only piecewise continuous or sometimes not even a classical function. The idea is as follows: given an ODE of the form $ Lu = f $, where $ L $ is some differential operator, one can transform the original ODE of the unknown function $ u $ to an algebraic equation $ \mathcal{L} \{Lu\} = \mathcal{L}\{f\} $ in the $ s $-domain for the Laplace transform $ U = \mathcal{L}\{u\} $ by using the Laplace transform. We'll see that this equation is generally easier to solve than the original ODE since it is an algebraic equation. Once we obtain a solution $ U $ in the $ s $-domain, we can obtain a solution $ u $ to the original ODE by "undoing" $ \mathcal{L} $, i.e. by taking the inverse Laplace transform $ \mathcal{L}^{-1} $. 

5. The Laplace transform is a special type of an *integral transform*. In general, they are of the form 
$$
T\{f\}(s) = \int_{D} K(s,t) f(t) \; dt, 
$$
where $f: \mathbb{F} \to \mathbb{F}$ is the input function, $K: \mathbb{F} \times \mathbb{F} \to \mathbb{F}$ is the *kernel* of the integral transform, $D \subseteq \mathbb{F}$ and usually $\mathbb{F} = \mathbb{R}$ or $\mathbb{F} = \mathbb{C}$. The Laplace transform is a special case of this where $K(s,t) = e^{-st}$. Other examples of integral transforms include the Fourier transform, the Mellin transform, the Hankel transform, etc. See the discussion [here](https://en.wikipedia.org/wiki/Integral_transform) for more details. 

### Sufficient conditions for existence

As we've pointed out in the last section, the Laplace transform of a function $f$ may not be well-defined. Here we give sufficient conditions on the function $f(t)$ so that the Laplace transform is well-defined for certain values of $s$. 

We will show that if a function $f$ is 
1. piecewise continuous
2. does not grow faster than an exponential $e^{ct}$ for large $t$, 

then its Laplace transform exists for $s > c$. To make this precise we need the following definitions.
> **Definition (Piecewise continuous)**
> 
> A function $f:\mathbb{R} \to \mathbb{R}$ is said to be *piecewise continuous* on a bounded interval $[a,b]$ if it is continuous on $[a,b]$ except at finitely many points (of which you see jump discontinuities). $f$ is said to be piecewise continuous on $[0,\infty)$ if it is piecewise continuous on every bounded subinterval of $[0,\infty)$. 
> 
> We can also define analogously the notion of piecewise smooth functions, which are functions for which the function is smooth (smooth meaning that all derivatives of the function are well-defined and continuous) restricted to the interior of each subinterval on which the function is continuous.

> **Definition (Exponential order)**
> 
> A function $f$ is said to be of *exponential order* if there exist constants $c, M, T > 0$ such that 
> $$
> \left| f(t) \right| \le M e^{ct}
> $$
> for all $t > T$.

One can easily check that polynomial functions and bounded functions are both functions of exponential order. With this definition we can deduce the following theorem.

> **Theorem**
> 
> Suppose $f$ is piecewise continuous on $[0,\infty)$ and is of exponential order with 
> $$
> \left| f(t) \right| \le M e^{ct}
> $$
> for all $t > T$. Then its Laplace transform $\mathcal{L}\{f\}$ exists for all $s > c$.

### Linearity of the Laplace transform

Recall the definition of a **linear operator**: if $X,Y$ are two vector spaces, an operator $T : X \to Y$ is said to be **linear** if for any $x,y \in X$ and any scalars $\alpha, \beta \in \mathbb{F}$, we have 
$$
T(\alpha x + \beta y) = \alpha T(x) + \beta T(y).
$$
If $T$ is an invertible linear operator, then its inverse $T^{-1} : Y \to X$ is also linear.

We note here that the Laplace transform is an example of a linear operator (but we're purposely vague here of the underlying vector spaces): for any two appropriate functions $f,g : [0,\infty) \to \mathbb{R}$ and any two real numbers $\alpha, \beta$, we have
$$
\mathcal{L} \{ \alpha f + \beta g \} = \alpha \mathcal{L} \{f \} + \beta \mathcal{L} \{ g\}.
$$


### The Laplace transform of $t^n$

> **Proposition**
> If $f(t) = t^n$ for some integer $n \in \mathbb{Z}_{\ge 0}$, then
> $$\mathcal{L}\{t^n\} (s) = \frac{n!}{s^{n+1}}, \; s > 0.$$

We will justify this by performing a slightly more general calculation below. 

### The Gamma function

> **Definition**
> We define the *gamma function* $\Gamma: (0,\infty) \to \mathbb{R}$ via 
> $$\Gamma(\alpha) := \int_0^\infty x^{\alpha-1}e^{-x} \; dx.$$

**Remark:** It is crucial to assume $\alpha >0$, so that the integral above converges.

**Remark:** $\Gamma(1) = \int_0^\infty e^{-x} \; dx = -e^{-x} \big\rvert_0^\infty = 1$.

> **Proposition:** $\Gamma(\alpha +1) = \alpha \Gamma(\alpha), \; \alpha > 0.$

**Justification:** Using integration by parts, and the fact that $\alpha > 0$ (exercise: where do we use it?)
$$
\begin{aligned}
\Gamma(\alpha +1) &= \int_0^\infty x^{\alpha}e^{-x} \; dx \\
&= -x^{\alpha} e^{-x} \bigg\rvert_0^\infty + \alpha \int_0^\infty e^{-x} x^{\alpha - 1} \; dx \\
&= -\left( \lim_{x \to \infty} x^\alpha e^{-x} - 0\right) + \alpha \Gamma(\alpha)\\
&= \alpha \Gamma(\alpha).
\end{aligned}
$$

> **Proposition**
> If we have a positive integer $n \in \mathbb{Z}^+$ then $\Gamma(n+1) = n!.$

**Justification:** By the previous proposition, 
$$
\Gamma(n+1) = n \Gamma(n) = n(n-1) \Gamma(n-1) = \cdots = n(n-1)\cdots 2 \cdot \Gamma(1) = n!.
$$

**Remark:** The previous proposition can be interpreted as saying that the gamma function can be viewed as the continuous analogue of the factorial function $n! := n(n-1)(n-2) \cdots 1$ for positive integers $n$.

Now we state the continuous analogue of the Laplace transform of $t^n$.

> **Proposition**
> For $\alpha > -1$, $\mathcal{L}\{ t^\alpha \} (s) = \frac{\Gamma(\alpha+1)}{s^{\alpha+1}}, \; s > 0.$

**Remark:** When $\alpha \ge 0$ is an integer this is equivalent to the Laplace transform of $t^n$ by the previous proposition.

**Justification:** We calculate:
$$
\begin{aligned}
\mathcal{L}\{t^\alpha\}(s) &= \int_0^\infty e^{-st} t^{\alpha} \; dt \\
(x=st, \frac{dx}{dt} = s)\quad &= \int_0^\infty e^{-x} (s^{-1} x)^{\alpha}\;  s^{-1}dx \\
&= s^{-\alpha -1} \int_0^\infty e^{-x} x^{\alpha} \; dx \\
&= \frac{1}{s^{\alpha +1} } \Gamma(\alpha +1).
\end{aligned}
$$

> **Lemma:** $\Gamma(1/2) = \sqrt{\pi}.$

**Justification:** We calculate
$$
\begin{aligned}
\Gamma \left( \frac{1}{2} \right) &= \Gamma \left( - \frac{1}{2} + 1\right) \\
&= \int_0^\infty e^{-x} x^{-1/2} \; dx \\
\left(\substack{u = \sqrt{x} \\ \frac{du}{dx} = 1/2 x^{-1/2}} \right) &= \int_0^\infty 2 e^{-u^2} \; du \\
&= \int_{-\infty}^\infty e^{-u^2} \; du.
\end{aligned}
$$
Then
$$
\begin{aligned}
\Gamma \left( \frac{1}{2} \right)^2 &= \left( \int_{-\infty}^\infty e^{-x^2} \; dx \right) \left( \int_{-\infty}^\infty e^{-y^2} \; dy \right) \\
&=  \int_{-\infty}^\infty   \int_{-\infty}^\infty e^{-(x^2 + y^2)} \; dx dy.
\end{aligned}
$$
Now we convert to polar coordinates. Mapping $(x,y) \mapsto (r,\theta)$ with $r^2 = x^2 + y^2, dx dy = r dr d\theta$,
$$
\begin{aligned}
\Gamma \left( \frac{1}{2} \right)^2 = \int_{-\infty}^\infty  \int_{-\infty}^\infty e^{-(x^2 + y^2)} \; dx dy &= \int_{0}^{2\pi}  \int_{0}^\infty e^{-r^2} r \; dr d\theta \\
&= 2\pi \cdot \frac{e^{-r^2}}{-2} \bigg\rvert_0^\infty \\
&= \pi.
\end{aligned}
$$
Therefore $\Gamma(1/2) = \sqrt{\pi}$.

**Remark:** As a consequence of this lemma we can calculate the Gamma function of all multiples of $1/2$ (the even multiples are just natural numbers). For example, 
$$
\Gamma \left( \frac{11}{2} \right) = \frac{9}{2} \cdot \frac{7}{2} \cdot \frac{5}{2} \cdot \frac{3}{2} \cdot \frac{1}{2} \cdot \Gamma \left( \frac{1}{2} \right) = \frac{9 \cdot 7 \cdot 5 \cdot 3 \cdot 1}{2^5} \sqrt{\pi}.
$$

## Inverse Laplace transforms

Next we examine how one can define the inverse of the Laplace transform, which maps a function from the frequency domain to the time domain. First we state a relevant proposition. 

> **Proposition**
> Suppose $f,g: [0,\infty) \to \mathbb{R}$ are piecewise continuous and of exponential order. Then $F = \mathcal{L}\{f\}, G = \mathcal{L}\{g\}$ are well-defined for $s > c$ for some constant $c \in \mathbb{R}$. If $F(s) = G(s)$ for all $s > c$, then $f(t) = g(t)$ for all $t \in [0, \infty)$ for which both $f,g$ are continuous at $t$. 

What this proposition is saying is that two piecewise continuous functions of exponential order with the same Laplace transform can only differ at finitely many points of discontinuity. Practically speaking, this means that the inverse of the Laplace transform is "essentially" unique. Making this mathematically precise is beyond the scope of this course, so we simply assume that if two piecewise continuous functions of exponential order have the same Laplace transform, then they must be the same function. This motivates the next definition.

> **Definition**
> Suppose $f: [0,\infty) \to \mathbb{R}$ is a continuous function of exponential order, and define $F = \mathcal{L}\{f\}$, which exists for $s > c$ for some $c \in \mathbb{R}$. The *inverse Laplace transform* is the operator that maps $F$ back to $f$, i.e. $\mathcal{L}^{-1}\{F\} = f$. 

**Remark:** The inverse Laplace transform is also linear:
$$
\mathcal{L}^{-1}\{\alpha F + \beta G\} = \alpha \mathcal{L}^{-1}\{F\} + \beta \mathcal{L}^{-1}\{G\}.
$$

**Remark:** As with the forward Laplace transform, we adopt an abuse of notation and write $\mathcal{L}^{-1} \{ F(s) \}$ even though $\mathcal{L}^{-1} \{ F(\cdot) \}$ is more appropriate. 


# Important Laplace and inverse transforms

In this subsection we list the Laplace and inverse Laplace transforms that we will often use.

1. 
    $$
    \mathcal{L} \{ 1\} = \frac{1}{s}, \; s > 0
    $$

2. 
    $$
    \mathcal{L} \{ t^\alpha\} = \frac{\Gamma(\alpha + 1)}{s^{\alpha +1}},  \; \alpha > -1.
    $$

3. 
    $$
    \mathcal{L} \{ e^{a t} \} = \frac{1}{s-a}, \; s > a
    $$

4. 
    $$
    \mathcal{L} \{ \sin (kt) \} = \frac{k}{s^2 + k^2}, \; s > 0
    $$

5. 
    $$
    \mathcal{L} \{ \cos( kt) \} = \frac{s}{s^2 + k^2}, \; s > 0.
    $$

With this we can write down a corresponding set of inverse transforms:

1. 
    $$
    \mathcal{L}^{-1} \left\{ \frac{1}{s} \right\} = 1, \; t \ge 0
    $$

2. 
    $$
    \mathcal{L}^{-1} \left\{ \frac{\Gamma(\alpha + 1)}{s^{\alpha +1}} \right\} = t^{\alpha}, \; t \ge 0
    $$

3. 
    $$
    \mathcal{L}^{-1} \left\{  \frac{1}{s-a} \right \} =e^{at}, \; t \ge 0
    $$

4. 
    $$
    \mathcal{L}^{-1} \left\{\frac{k}{s^2 + k^2} \right \} = \sin (kt), \; t \ge 0
    $$

5. 
    $$
    \mathcal{L}^{-1}  \left\{ \frac{s}{s^2 + k^2} \right \} = \cos(kt), \; t \ge 0
    $$

Next we discuss how derivatives of piecewise smooth functions satisfying appropriate growth conditions transform under the Laplace transform. 

> **Proposition (Laplace transforms of derivatives)**
> Let $y$ be a piecewise smooth function of exponential order and assume that all of its derivatives are continuous and also of exponential order. Denote its Laplace transform $\mathcal{L}\{ y \}$ by $Y$. Then 
> $$
> \mathcal{L} \{ y^{(n)} (t) \} = s^n Y(s) - s^{n-1}y(0) - s^{n-2} y'(0) - ... - s y^{(n-2)}(0) - y^{(n-1)} (0), \; \text{for all} \; s > c
> $$
> for some constant $c \in \mathbb{R}$.
> In particular,
> $$
> \mathcal{L} \{ y' \} = sY(s) - y(0)
> $$
> $$
> \mathcal{L} \{ y'' \} = s^2Y(s) - s y(0) - y'(0)
> $$
> for all $s > c$. 

Similarly, differentiation in the frequency domain corresponds to multiplying by polynomial factors in the time domain.

> **Proposition**
> Suppose $f:[0,\infty) \to \mathbb{R}$ is a function for which its Laplace transform is smooth for $s > c$ for some $c \in \mathbb{R}$. Then 
> $$
> \mathcal{L} \{ t^n f(t) \} = (-1)^n \frac{d^n}{ds^n} F(s), \; s > c
> $$
> for any integer $n \ge 0$.

**Remark:** This is saying that multiplying by powers of $t$ in the $t$-domain correspond to differentiating in the $s$-domain.

> **Corollary**
> $$
> \mathcal{L}^{-1} \left\{ \frac{d^n}{ds^n} F(s) \right\} = (-1)^n t^n f(t).
> $$

**Remark:** We can use the previous proposition to compute
$$
\mathcal{L}^{-1} \left\{ \frac{1}{(s^2 + 1)^2} \right\}.
$$
First notice that 
$$
\frac{d}{ds} \mathcal{L} \{ \cos(t) \} = \frac{d}{ds} \left( \frac{s}{s^2 + 1} \right) = \frac{1-s^2}{(s^2+1)^2},
$$
and
$$
\frac{d}{ds} \mathcal{L} \{ \sin(t) \} = \frac{d}{ds} \left( \frac{1}{s^2 + 1} \right) = \frac{-2s}{(s^2+1)^2}.
$$
Then
$$
\begin{split}
 \frac{1}{(s^2 + 1)^2} &= \frac{s^2 + 1 - s^2}{(s^2+1)^2} \\
 &= \frac{1}{s^2+1} - \frac{s^2}{(s^2+1)^2} \\
 &= \frac{1}{s^2+1} - \frac{s^2 - 1 + 1}{(s^2+1)^2}  \\
 &= \frac{1}{s^2+1} - \frac{s^2 - 1}{(s^2+1)^2} - \frac{1}{(s^2+1)^2} \\
 \implies 2 \frac{1}{(s^2+1)^2} &= \frac{1}{s^2+1} + \frac{1-s^2}{(s^2+1)^2} \\
 \implies  \frac{1}{(s^2 + 1)^2} &= \frac{1}{2} \frac{1}{s^2+1}  +  \frac{1}{2}  \underbrace{\frac{1-s^2}{(s^2+1)^2}}_{=\frac{d}{ds} \mathcal{L}\{\cos(t)\} }.
 \end{split}
$$
This implies that 
$$
\begin{split}
\begin{split}
\mathcal{L}^{-1} \left\{ \frac{1}{(s^2 + 1)^2} \right\} &= \frac{1}{2} \mathcal{L}^{-1} \left\{ \frac{1}{s^2+1}  \right\}  + \frac{1}{2} \mathcal{L}^{-1} \left \{   \frac{1-s^2}{(s^2+1)^2}\right\} \\
&= \frac{1}{2} \sin(t) - \frac{1}{2} t \cos(t).
\end{split}
 \end{split}
$$
