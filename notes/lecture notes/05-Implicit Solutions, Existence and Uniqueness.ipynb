## Implicit Solutions 

Recall from the last notebook that in the process of solving a separable equation, we first at a relation between the dependent and independent variable first, and then we try to write the dependent variable as a function of the independent variable. However, not all relations can be easily solved for the dependent variable. For this reason, we introduce the notion of an **implicit solution**.

> **Definition**:
> A function $G: I \times \mathbb{R} \to \mathbb{R}$ defined via $G(x,y) = 0$ defines an **implicit solution** to a first order differential equation over an interval $I$ if there exists a $C^1$ function $f: I \to \mathbb{R}$ such that $G(x,f(x)) = 0$ for all $x \in I$. 

**Remark:**
We will assume in this class that if an identity of the form $G(x,y) = 0$ can be derived in the process of manipulating equations, then there exists an interval $I$ and a differentiable function $y$ for which $G(x,y(x)) = 0$ for all $x \in I$. Investigating when a relation of the form $G(x,y) = 0$ defines a differentiable function requires the *implicit function theorem*, which is outside the scope of this course. 

> **Example:**
> Consider the differential equation, 
> $$y'(x) = -\frac{x}{y(x)}, \; x \in I.$$
> We claim that the relation 
> $$G(x,y) = x^2 + y^2 - r^2 = 0, \; (x,y) \in \mathbb{R}$$
> for any constant $r \neq 0$ defines an implicit solution to the equation. We note that via implicit differentiation, 
> $$2x + 2 y(x) y'(x) = 0, x \in (-r,r),$$
> therefore 
> $$y'(x) = -\frac{x}{y(x)}, \; x \in (-r,r).$$
> We note that in this case it is possible to solve for $y$ explicitly, a solution $y: (-r,r) \to \mathbb{R}$ can either be defined via $y(x) = \sqrt{r^2 - x^2}$ or $y(x) = -\sqrt{r^2 - x^2}$. If an initial condition is given, then we choose the branch of the square root that satisfies the initial condition.

> **Example:**
> Consider the initial value problem 
> $$
>   \begin{cases}
>     (e^{2y(x)} - y(x)) \cos(x) y'(x) = e^{y(x)} \sin 2x, \; x \in \mathbb{R} \\
>     y(0) = 0.
>  \end{cases}
> $$
> We note that this is a separable equation: if $y$ is a solution, then on some interval $I$ we have 
> $$
>    \frac{e^{2y(x)} - y(x)}{e^{y(x)}} y'(x) = \frac{\sin 2x}{\cos x} = 2 \sin x, \;x \in I.
> $$
> Thus we have the identity 
> $$
>    \int (e^y - ye^{-y}) \; dy = 2 \int \sin x \; dx.
> $$
> This leads us to the implicit relation
> $$
>    G(x,y) = e^{y} + y e^{-y} + e^{-y} + 2 \cos x + C = 0, \; (x,y) \in \mathbb{R}^2.
> $$
> If $y(0) = 0$, then we see that $C = -4$.


## Existence and Uniqueness of Solutions 

For a general initial value problem of the form 
$$
    \begin{cases}
    y'(t) = f(t,y(t)), \; t \in I \\
    y(t_0) = y_0,
    \end{cases}
$$
we say that
- a solution **exists** if there is at least one $C^1$ solution to the IVP
- a solution is **unique** if the following hold: if $y_1, y_2$ are two solutions to the IVP, then it must be the case that $y_1$ is equal to $y_2$ over $I$.

Two main questions we want to ask in this class are:
1. What conditions can we impose on $f$ to guarantee the existence of a solution? 
2. What additional conditions can we impose on $f$ to guarantee uniqueness?

In practice one also asks a third question:

3. If a unique solution exists, does the solution depend continuously on the initial condition?

If an IVP admits a unique solution and satisfies 3, then we say that the IVP is **well-posed**. This definition is given by Jacques Hadamard (1865-1963) in 1902. There are entire books written on studying well-possedness of ODEs and PDEs. For us, we will only write down a few sufficient conditions to guarantee existence and uniqueness.


The upshot of the discussion below is that 
- existence (locally) is guaranteed if $f$ is continuous on some rectangle $R$, though solutions may not be unique;
- uniqueness (locally) is guaranteed if both $f$ and $\frac{\partial f}{\partial y}$ are continuous on some rectangle $R$.

We note that these theorems are typically local in nature as the ODE can be nonlinear. 


Demanding an IVP to be well-posed is natural for physical reasons: if we model a physical system via an initial value problem and find that there are infinitely many solutions, there is an immediate question as to which of the infinitely many solutions accurately models the physical system. 

We'll also see that uniqueness has some very important consequences, one of which is related to the structure of linear differential equations.

> **Example (Catastrophic failure of uniqueness)**:
> Consider the initial value problem 
> $$
>     \begin{cases}
>      y'(t) = \sqrt{|y(t)|}, & t \in \mathbb{R} \\
>      y(0) = 0.
>     \end{cases}
> $$
> One can verify that for any constant $a \ge 0$, any function $y: \mathbb{R} \to \mathbb{R}$ of the form 
> $$
>     y(t) = \begin{cases}
>       \frac{1}{4}(t-a)^2, & t > a \\
>       0, & t \le a.
>     \end{cases}
> $$
> is a solution to the initial value problem. First, we note that $y(0)= 0$ since $a \ge 0$. Then, we note that
> $$
>   y'(t) = \begin{cases}
>      \frac{1}{2} (t-a), &  t > a \\
>   0, & t \le a
> \end{cases} 
> $$
> $$
>   \sqrt{|y(t)|} = \begin{cases}
> \sqrt{|\frac{1}{4} (t-a)^2|},&  t > a \\
> \sqrt{|0|},&  t \le a 
> \end{cases}  = \begin{cases}
> \frac{1}{2} |t-a|,&  t > a\\
> 0, & t \le a
> \end{cases} = \begin{cases}
> \frac{1}{2} (t-a), & t > a \\
> 0, & t \le a.
> \end{cases}
> $$
> Since $y'(0) = 0$, we also see that $y'$ is continuous, therefore $y: J \to \mathbb{R}$ is a solution for any $a \ge 0$ for $J = \mathbb{R}$.

## An existence theorem

Next we state the main theorems of this section.

> **Theorem (Peano existence):** Consider the initial value problem 
> $$
> \begin{cases}
> y'(t) = f(t,y(t)), \; t \in \mathbb{R} \\
> y(t_0) = y_0.
> \end{cases}  \quad \quad (\text{IVP})
> $$
> Let $R \subseteq \mathbb{R}^2$ be an open rectangular region of the form
> $$
> R: \{ (t,y) \in \mathbb{R}^2 : a < t < b,  c < y < d, \; a,b,c,d \in \mathbb{R} \} 
> $$
> containing the point $(t_0, y_0)$ as specified above. If $f: R \to \mathbb{R}$ is continuous, then there exists an open subinterval $I$ of $(a,b)$ containing $t_0$ and a differentiable function $y: I \to \mathbb{R}$ satisfying the initial value problem (IVP).

> **Theorem (Picard-LindelÃ¶f):** Let $R$ be an open rectangular region of the form
> $$
> R: \{ (t,y) \in \mathbb{R}^2 : a < t < b,  c < y < d, \; a,b,c,d \in \mathbb{R} \} 
> $$
> containing the point $(t_0, y_0)$ as specified above. If $f:  R \to \mathbb{R}$ and $\frac{\partial f}{\partial y}:  R \to \mathbb{R}$ are continuous, then there exists an open subinterval $I$ of $(a,b)$ containing $t_0$ and a **unique** differentiable function $y: I \to \mathbb{R}$ satisfying the initial value problem (IVP).


We note that in the example above, the function $f$ is  
$$
    f(t, y) = \sqrt{|y|}, \; (t,y) \in \mathbb{R}^2.
$$
Taking a partial derivative w.r.t. to the second variable gives us 
$$
    \frac{\partial f}{\partial y}(t,y) = \frac{1}{2 \sqrt{|y|}} \frac{y}{|y|}, \; t \in \mathbb{R}, y \neq 0.
$$
We note that since $\frac{\partial f}{\partial y}$ is not defined at $y = 0$, it is not continuous in any rectangle containing $(0,0)$. Therefore the uniqueness theorem does not apply.

## Existence and Uniqueness for First Order Linear ODEs

Next we state a version of the existence and uniqueness theorem for first order linear ODEs.

> **Theorem:**
> Consider an initial value problem of the form 
> $$
>     \begin{cases}
>         a_1(t) y'(t) + a_0(t) y(t) = f(t), \; t \in I \\
>         y(t_0) = y_0.
>     \end{cases}
> $$
> If $a_1, a_0, f$ are continuous on the interval $I$ and $a_1(t) \neq 0$ for all $t \in I$, then there exists a solution to the initial problem on the interval $I$ and it is unique.

**Remark:** Note that unlike the more general theorems stated above, this theorem guarantees the existence and uniqueness of a global solution, whereas the previous theorems only guarantee the existence and uniqueness of a local solution.

**Remark:**
We note that this only applies when we consider an interval for which $a_1 \neq 0$. We will see through examples that strange things can happen at values $t$ for which $a_1(t) = 0$. 


## Consequences of Uniqueness

The following is an immediate consequence of the uniqueness theorem.

> **Proposition (Different solution curves cannot cross):** 
> Suppose we are given 
> $$
> y'(t) = f(t,y(t)), \; t \in I
> $$
> and solutions to this ODE are unique on $I$ (suffices to assume that $f: I \times \mathbb{R} \to \mathbb{R}$ and $\frac{\partial f}{\partial y}: I \times \mathbb{R} \to \mathbb{R}$ are continuous). If $y_1, y_2$ are two different solutions satisfying the equation on some interval $J \subseteq I$, in the sense that there exists at least one $t \in J$ such that $y_1(t) \neq y_2(t)$, the solutions curves $y_1, y_2$ cannot cross paths in $J$. In other words, $y_1(t) \neq y_2(t)$ for all $t$ in $J$. 

**Justification:**
Suppose by contradiction that $y_1(t^*) = y_2(t^*) = y^*$ for some $t^*$ in $J$. Then $y_1, y_2$ are both solutions to the IVP
$$
\begin{cases}
y'(t) = f(t,y(t)), \; t \in J \\
y(t^*) = y^*.
\end{cases}
$$
over some interval containing $t^*$. By uniqueness, we must have $y_1 \equiv y_2$ on this interval, which contradicts the assumption that $y_1, y_2$ are two different solution curves.
