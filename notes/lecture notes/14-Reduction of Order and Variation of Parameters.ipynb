## Reduction of order

Recall that the general form of a 2nd order linear homogeneous differential equation is
$$
y''(t) + p(t)y'(t) + q(t)y(t) = 0, \; t \in I
$$
where $I$ is an interval and $p, q: I \to \mathbb{R}$ are assumed to be continuous. The general solution to this equation is given by $y: I \to \mathbb{R}$ given by
$$
y(t) = c_1 y_1(t) + c_2 y_2(t)
$$
where $y_1, y_2: I \to \mathbb{R}$ are linearly independent and $c_1, c_2$ are arbitrary constants. If $p,q: I \to \mathbb{R}$ are constant functions, then $y_1, y_2$ can be identified explicitly via analyzing the characteristic equation associated to the ODE. If they are not, then there is no simple mechanism for identifying $y_1, y_2$. In this section we go over a method that allows us to recover one of them given that the other one is known.

### The method 
Suppose $y_1 : I \to \mathbb{R}$ is known. We know that any function of the form $c y_1$ is linearly dependent on $y_1$, so in order to find a second linearly independent solution, we look for a solution of the form $y_2: I \to \mathbb{R}$ given by $y_2(t) = u(t) y_1(t)$ where $u: I \to \mathbb{R}$ is a differentiable function to be determined. The idea is to use this as an ansatz and substitute it into the ODE. We first calculate
$$
y_2'(t) = u'(t) y_1(t) + u(t) y_1'(t), \; y_2''(t) = u''(t) y_1(t) + 2 u'(t) y_1'(t) + u(t) y_1''(t), \; t \in I.
$$
Then if $y_2$ is a solution, we must have
$$
y_2'' + p y_2' + q y_2 = \left(u'' y_1 + 2 u' y_1' + u y_1''\right) + p (u' y_1 + u y_1') + q \left( u y_1\right) = 0.
$$
Note that we can simplify this by using the fact that $y_1$ is a solution:
$$
\begin{align*}
0 &=  \left(u'' y_1 + 2 u' y_1' + u y_1''\right) + p (u' y_1 + u y_1') + q \left( u y_1\right) \\
&= u'' \left( y_1 \right) + u' \left( 2 y_1' + p y_1   \right) + u \underbrace{\left( y_1'' + p y_1' + q y_1 \right)}_{= 0}
\end{align*}
$$
Note that if we define $v: I \to \mathbb{R}$ via $v(t) = u'(t)$, then the equation above is a first order linear equation in $v$:
$$
y_1(t) v'(t) + (2 y_1'(t) + p y_1(t)) v(t) = 0.
$$
This is an equation that can be solved using the method of integrating factors or separation of variables. Once $v$ is identified, $u$ can be recovered by direct integration, and subsequently $y_2$. By construction, $y_2$ is a solution to the ODE, and as long as $y_1, y_2$ are linearly independent, we have recovered the general solution to the ODE.

> **Example:**
> Consider the equation 
> $$
> t^2 y''(t) + 2t y'(t) - 2 y(t) = 0, \; t > 0.
> $$
> One can verify directly that $y_1: (0, \infty) \to \mathbb{R}$ given by $y_1(t) = t$ is a solution. Here we use the reduction of order method to identify a second linearly independent solution. We look for a solution $y_2: I \to \mathbb{R}$ of the form $y_2(t) = u(t) t$ where $u: (0, \infty) \to \mathbb{R}$ is a differentiable function to be determined. We calculate
> $$
> y_2'(t) = t u'(t) + u(t), \; y_2''(t) = t u''(t)  + 2 u'(t), \; t > 0.
> $$
> Then if $y_2$ is a solution, we must have
> $$
> \begin{align*}
> 0 &= t^2 y_2''(t) + 2t y_2'(t) - 2 y_2(t) \\
> &= t^2 \left( t u''(t) + 2 u'(t) \right) + 2t \left(t  u'(t)  + u(t) \right) - 2 t u(t) \\
> &= t^3 u''(t) + 4 t^2 u'(t) + 2 t u(t) - 2 t u(t)  \\
> &= t^3 u''(t) + 4 t^2 u'(t), \; t > 0.
> \end{align*} 
> $$
> Note that the $u$ terms canceled out - this is an indication that the algebra is correct. Then $v: I \to \mathbb{R}$ defined via $v = u'$ satisfies the first order equation 
> $$
> v'(t) + \frac{4}{t} v(t) = 0, \; t > 0.
> $$
> Using the method of integrating factors, we find that $v(t) = c t^{-4}$ for some constant $c \in \mathbb{R}$. Then $u(t) = \alpha t^{-3} + \beta$ for arbitrary constants $\alpha, \beta \in \mathbb{R}$, which implies that $y_2(t) = u(t) y_1(t) = \alpha t^{-2} + \beta t$. This shows that the general solution to the homogeneous equation is given by
> $$
> y(t) = c_1 t + c_2 t^{-2}, \; t > 0,
> $$
> where $c_1, c_2$ are arbitrary.

## Variation of parameters

Reduction of order is a method for recovering the general solution to a homogeneous second order linear equation, given that one of the solutions is known. Variation of parameters is an extension of the idea applied to inhomogeneous equations. 

Consider a general second order linear inhomogeneous equation of the form
$$
y''(t) + p(t) y'(t) + q(t) y(t) = f(t), \; t \in I
$$
and $\{ y_1, y_2 \} \subseteq C^2(I; \mathbb{R})$ is a fundamental set of solutions to the homogeneous equation that is already known. Since any linear combination $c_1 y_1 + c_2 y_2$ is a homogeneous solution, we consider an ansatz $y_p: I \to \mathbb{R}$ of the form
$$
y_p(t) = u_1(t) y_1(t) + u_2(t) y_2(t)
$$
where $u_1, u_2: I \to \mathbb{R}$ are differentiable functions to be determined. The idea is to substitute this into the inhomogeneous equation and determine $u_1, u_2$ such that the equation is satisfied. Note that for reduction of order, we identified one constraint (a first order ODE) for one unknown. Here we aim to identify two constraints for two unknowns.

We first compute 
$$
y_p' = u_1' y_1 + u_2' y_2 + u_1 y_1' + u_2 y_2' = (u_1 y_1' + u_2 y_2') + \underbrace{\textcolor{red}{(u_1' y_1 + u_2' y_2)}}_{= (*)},
$$
We note that if $(*) = 0$, then the second derivatives of $u_1, u_2$ would not appear in $y_p''$. It turns out this makes the algebra works out nicely, so this is one of the constraints we will impose. We then compute
$$
y_p'' = (u_1 y_1'' + u_2 y_2'') + (u_1' y_1' + u_2' y_2') = u_1' y_1' + u_2' y_2' - p(u_1 y_1' + u_2 y_2') - q(u_1 y_1 + u_2 y_2) =  u_1' y_1' + u_2' y_2' - py_p' - q y_p,
$$
using the fact that $y_i'' = - py_i' - q y_i$ for $i = 1, 2$. Then it follows that
$$
y_p'' + p y_p' + q y_p = u_1' y_1' + u_2' y_2'.
$$
Therefore for $y_p$ to be a particular solution to the inhomogeneous equation, we require 
$$
(**) \quad \begin{cases}
u_1' y_1 + u_2' y_2 = 0 \\
u_1' y_1' + u_2' y_2' = f,
\end{cases} \quad t \in I.
$$
Note that if we multiply the first equation by $y_1'$ and the second equation by $y_1$, we obtain
$$
\begin{cases}
u_1' y_1 y_1' + u_2' y_1' y_2  = 0 \\
u_1' y_1 y_1' + u_2' y_1 y_2'  = f y_1,
\end{cases} \quad t \in I.
$$
Subtracting the first equation from the second, we obtain
$$
u_2'(y_1 y_2' - y_1' y_2) = f y_1.
$$
Recall that the Wronskian $W: I \to \mathbb{R}$ is defined by $W(t) = y_1(t) y_2'(t) - y_1'(t) y_2(t)$. On the homework we have shown that if $y_1, y_2$ are linearly independent solutions, then $W \neq 0$ over $I$. From this it follows that 
$$
u_2' = \frac{f y_1}{W} \implies u_2(t) = \int \frac{f(t) y_1(t)}{W(t)} dt.
$$
From (**), we can also multiply the first equation by $y_2'$ and the second equation by $y_2$ to obtain
$$
\begin{cases}
u_1' y_1 y_2' + u_2' y_2 y_2'  = 0 \\
u_1' y_1' y_2 + u_2'y_2 y_2'   = f y_2,
\end{cases} \quad t \in I.
$$
and subtracting the second equation from the first, we obtain
$$
u_1'(y_1 y_2' - y_1' y_2) = -f y_2.
$$
From this it follows that
$$
u_1' = - \frac{f y_2}{W} \implies u_1(t) = - \int \frac{f(t) y_2(t)}{W(t)} dt.
$$
This shows that a candidate particular solution $y_p: I \to \mathbb{R}$ is given by
$$
y_p(t) = - y_1(t) \int \frac{f(t) y_2(t)}{W(t)} dt + y_2(t) \int \frac{f(t) y_1(t)}{W(t)} dt.
$$
This is the variations of parameters formula, and you are asked to verify that the candidate solution given above is indeed a solution to the inhomogeneous equation.

The variation of parameter formula shows that as long as a set of independent homogeneous solutions is known, one can construct a particular solution via the variation of parameters formula in terms of the homogeneous solutions and the inhomogeneous term. After the particular solution is identified, the general solution to the inhomogeneous equation is given by the sum of the particular solution and the general solution to the homogeneous equation.
