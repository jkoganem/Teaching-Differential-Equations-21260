In this notebook we go over special classes of first order equations that can be solved by *substitution methods*. 

## Bernoulli differential equations

Bernoulli differential equations are first order differential equations of the form 
$$
y'(x) + P(x)y(x) = Q(x)(y(x))^{\alpha}, \; x \in \R,
$$
with $\alpha \neq 0, 1$.
  
Since $\alpha \neq 0,1$, the equation is nonlinear and the method of integrating factors cannot be applied.

One can show that the substitution 
$$
v(x) = (y(x))^{1-\alpha}, \; x \in I
$$
for some interval $I$ reduces the equation to a first order linear equation.

> **Example**
> 
> Consider the differential equation 
> $$
> y'(x) + \frac{6y(x)}{x} = 3 (y(x))^{4/3}, \; x > 0.
> $$
> This is a Bernoulli equation with $\alpha = 4/3$. Suppose $y$ is a solution and there exists an interval $I \subseteq (0,\infty)$ for which $y(x) \neq 0$ for all $x \in I$, and consider the function 
> $$
> v(x) = (y(x))^{-1/3}, \; x \in I.
> $$
> Then 
> $$
> v'(x) = -\frac{1}{3} (y(x))^{-4/3}y'(x), \; x \in I.
> $$
> Notice that by dividing both sides of the original equation by $y^{4/3}$ on $I$, we may also deduce that 
> $$
> y'(x)(y(x))^{-4/3} + \frac{6}{x}(y(x))^{-1/3} = 3, \; x \in I.
> $$
> Then we see that $v$ satisfies the differential equation 
> $$
> v'(x) - \frac{2}{x} v(x) = -1, \; x \in I.
> $$
> This is a first order linear equation for $v$. We may then choose an integrating factor to be 
> $$
> \mu(x) = \exp \left( \int \frac{-2}{x} \; dx \right) = \exp \ln |x|^{-2} = x^{-2}, \; x \in I.
> $$
> Thus if $v$ is a solution, then 
> $$
> \frac{d}{dx}[x^{-2} v(x)] = -x^{-2}, \; x \in I.
> $$
> Thus 
> $$
> v(x) = x + Cx^2, \; x \in I,  C \in \R.
> $$
> Therefore a candidate solution to the original equation is 
> $$
> y(x) = \frac{1}{(x + Cx^2)^3}, \; x \in I, C \in \R.
> $$
> If we solve a corresponding initial value problem, the maximum interval of existence $J$ will depend on the initial conditions specified, since $x + Cx^2$ can be equal to zero.


## First order equations with homogeneous functions 

Consider first order differential equations of the form 
$$
y'(x) = F\left(\frac{y(x)}{x} \right), \; x > 0,
$$
where $F: \mathbb{R} \to \mathbb{R}$ is assumed to be continuous. 

A function that is a function of $y/x$ is usually referred to as a *homogeneous function*. For this reason these equations are sometimes referred to as *homogeneous differential equations*, though this terminology is inconsistent with the notion of homogeneous equations that we have introduced previously. Since in this class linear equations are emphasized more than nonlinear equations, I will instead refer to these specialized equations as *differential equations with homogeneous functions*.

### Substitution 
Consider the function $v: (0,\infty) \to \mathbb{R} $ defined via 
$$
v(x) = \frac{y(x)}{x}, \; x > 0.
$$
This is equivalent to 
$$
y(x) = x v(x), \; x > 0,
$$
and therefore
$$
y'(x) = v(x) + x v'(x), \; x > 0.
$$
We may then see that if $y$ solves the original equation, then $v$ solves the separable differential equation. 
$$
v'(x) = \frac{1}{x} \left( F(v(x)) - v(x) \right), \; x > 0.
$$

> **Example**
> 
> Consider the differential equation 
> $$
> 2x y(x)y'(x) = 4x^2 + 3(y(x))^2, \; x > 0.
> $$
> If $y$ is a solution to this equation, then there exists an interval $I \subseteq (0,\infty)$ for which $y(x) \neq 0$ for all $x \in I$. Thus 
> $$
> y'(x) = \frac{2x}{y(x)} + \frac{3}{2} \frac{y(x)}{x}, \; x \in I.
> $$
> Notice that the right hand side is a function of $y/x$.
> 
> Using the substitution 
> $$
> v(x) = \frac{y(x)}{x}, \; x \in I,
> $$
> we see that 
> $$
> y(x) = x v(x) \implies y'(x) = v(x) + x v'(x), x \in I.
> $$
> Thus $v$ satisfies the equation 
> $$
> v(x) + x v'(x) = \frac{2}{v(x)} + \frac{3}{2} v(x), \; x \in I.
> $$
> 
> We then see that $v$ satisfies 
> $$
> \frac{2v(x)}{(v(x))^2 + 4} v'(x) = \frac{1}{x}, \; x \in I,
> $$
> and therefore we have the identity 
> $$
> \int \frac{2v}{v^2 + 4} \; dx = \int \frac{1}{x} \; dx.
> $$
> Thus 
> $$
> \ln ((v(x))^2 + 4) = \ln |(v(x))^2 + 4| = \ln |x| + C, \; x \in I.
> $$
> 
> Since $x > 0$ if $x \in I, we find that 
> $$
> \frac{(y(x))^2}{x^2} +4 = C x, \; x \in I, C > 0.
> $$
> Therefore a candidate implicit solution is given by 
> $$
> (y(x))^{2} + 4x^2 = Cx^3, \; x \in I, C > 0.
> $$
> A candidate explicit solution would then be 
> $$
> y(x) = \pm \sqrt{Cx^3 - 4x^2}, \; x \in J, C > 0
> $$
> for some interval $J$.


## Exact differential equations
Recall that for certain differential equations, the solution $y$ can only be written implicitly as 
$$
F(x,y(x)) = C, \; x \in I
$$
where $I$ is some interval, $C$ is an arbitrary constant, and $F: \mathbb{R}^2 \to \mathbb{R}$ is $C^1$.

We note that differentiating this expression with respect to $x$ gives us 
$$
\frac{\partial F}{\partial x} (x,y(x)) + \frac{\partial F}{\partial y}(x,y(x)) \frac{dy}{dx}(x) = 0, \; x \in I.
$$
We can then write this equation in the form of 
$$
M(x,y(x)) + N(x,y(x)) y'(x) = 0.
$$

This shows that if we have a differential equation of the form
$$
M(x,y(x)) + N(x,y(x)) y'(x) = 0,
$$
and **if** there exists a function $F: \mathbb{R}^2 \to \mathbb{R}$ for which 
$$
\frac{\partial F}{\partial x}(x,y) = M(x,y) \; \text{and} \; \frac{\partial F}{\partial y}(x,y) = N(x,y) \; \text{for all} \; x \in I,
$$
then the identity 
$$
F(x,y(x)) = C, \; x \in I
$$
gives us an implicit solution to the equation. 

We refer to these equations as **exact differential equations**. We note that some textbooks adopt the shorthand differential notation and write the equation as $M dx + N dy = 0$.


Two immediate questions.

1. How can we determine if a differential equation is exact?
2. How can we recover the function $F$?


Recall that under mild conditions, we have symmetry of second derivatives:
$$
\frac{\partial}{\partial x \partial y} F = \frac{\partial}{\partial y \partial x} F.
$$
Therefore if an equation is exact, and 
$$
\frac{\partial F}{\partial x}(x,y) = M(x,y) \; \text{and} \; \frac{\partial F}{\partial y}(x,y) = N(x,y) \; \text{for all} \; x \in I,
$$
then it is necessary for 
$$
\frac{\partial M}{\partial y}(x,y) = \frac{\partial N}{\partial x} (x,y), \; x \in I.
$$

The previous calculation shows us that since it is necessary for $M_y = N_x$, if $M_y \neq N_x$ then the equation is not exact, and we cannot find a function $F$ for which $F_x = M$ and $F_y = N$. 


> **Example**
> Consider the exact differential equation
> $$
> (y(x))^3 + 3x (y(x))^2 y'(x) = 0, \; x \in \R.
> $$
> The function
> $$
> F(x,y) = xy^3, \; (x,y) \in \R^2
> $$
> gives us a general solution to the original equation on some interval $I$, as the general implicit solution is 
> $$
> x (y(x))^3 = C, \; x \in I
> $$
> and $C$ is an arbitrary constant.
> 
> In this example we have $M, N : \mathbb{R}^2 \to \mathbb{R}$ given by
> $$
> M(x,y) = y^3, \; N(x,y) = 3xy^2.
> $$
> Note that it is necessary for $M_y = N_x$, which we can see explicitly:
> $$
> \frac{\partial M}{\partial y}(x,y) = 3 y^2 \; \text{and} \; \frac{\partial N}{\partial x}(x,y) = 3y^2.
> $$

> **Example**
> Consider the differential equation 
> $$
> y(x) + 3x y'(x) = 0, \; x \in \R,
> $$
> which is obtained from the equation in the first example upon dividing through by $y^2$. Here we see that if we define $M,N: \mathbb{R}^2 \to \mathbb{R}$ via 
> $$
> M(x,y) = y, \; N(x,y) = 3x,
> $$
> then 
> $$
> \partial_y M(x,y) = 1, \; \partial_x N(x,y) = 3, \; (x,y) \in \R^2.
> $$
> Therefore the equation is not exact.
> 
> This example shows us that whether an equation is exact or not depends on the precise form in which it is written.

The next proposition gives us a criterion to check whether an equation is exact or not. 

> **Proposition**
> Let $R = (a,b) \times (c,d) \subset \R^2$ be an open rectangle, and suppose $M,N \in C^1(R; \mathbb{R})$. Then the differential equation 
> $$
> M(x,y(x)) + N(x,y(x)) y'(x) = 0, \; x \in (a,b)
> $$
> is exact in $R$ if and only if 
> $$
> \frac{\partial M}{\partial y} (x,y) = \frac{\partial N}{\partial x} (x,y) \; \text{for all} \; (x,y) \in R.
> $$
> That is, there exists a function $F$ defined on $R$ for which $F_x = M$ and $F_y = N$ if and only if $M_y = N_x$. 

### Recovering the function $F$

Note that if $M,N$ are given and $M_y = N_x$, then we can recover $F$ by using the fact that $F_x = M$ and $F_y = N$. First, note that for any function $g: \mathbb{R} \to \mathbb{R}$ that is a function of $y$, the function 
$$
F(x,y) = \int M(x,y) \; dx + g(y), \; (x,y) \in \R^2
$$
satisfies $F_x = M$. Here the notation $\int M(x,y) \; dx$ denotes an antiderivative of $M$ as a function of $x$, with $y$ fixed as a constant. Then, using $F_y = N$ we find that 
$$
N(x,y) = \frac{\partial}{\partial y} \int M(x,y) \; dx + g'(y),
$$
which is an identity for $g'$. We may then use this to recover $g$, which recovers $F$. 


> **Example**
> Consider the differential equation
> $$
> (6xy(x) - (y(x))^3) + (4(y(x)) + 3x^2 - 3x (y(x))^2) y'(x) =0 , \; x \in \R.
> $$
> To check whether this equation is exact we define $M,N: \mathbb{R}^2 \to \mathbb{R}$ via 
> $$
> M(x,y) = 6xy - y^3, \; N(x,y) = 4y+ 3x^2 - 3xy^2.
> $$
> Then we may calculate  
> $$
> \frac{\partial M}{\partial y}(x,y) = 6x - 3y^2, \; (x,y) \in \R^2
> $$
> and 
> $$
> \frac{\partial N}{\partial x} (x,y) = 6x - 3y^2, \; (x,y) \in \R^2.
> $$
> Therefore the equation is exact. To find $F$, we note that since we must have $F_x = M$, we must have 
> $$
> F(x,y) = \int (6xy - y^3)\; dx = 3x^2 y - xy^3 + g(y), \; (x,y) \in \R^2. 
> $$
> Here $\int \cdot \; dx$ refers to the antiderivative w.r.t. to the $x$ variable while holding $y$ fixed, so we treat $y$ as a constant when we integrate. 
> 
> Since we must also have $F_y =N$, we must also have 
> $$
> \frac{\partial F}{\partial y}(x,y) = 3x^2 - 3x y^2 + g'(y) = N(x,y) = 4y + 3x^2 - 3xy^2, \; (x,y) \in \R^2.
> $$
> This implies that we must have $g'(y) = 4y$ for $y \in \R$, therefore 
> $$
> F(x,y) = 3x^2y - xy^3 + 2y^2 + C, \; (x,y) \in \R^2.
> $$
> Thus the general implicit solution on some interval $I$ is 
> $$
> 3x^2y(x) - x(y(x))^3 + 2(y(x))^2 = C, \; x \in I
> $$
> where $C$ is arbitrary.
